```python
import matplotlib.patches as patches
import sys
from pathlib import Path

# posibles ubicaciones (para vos y para tu amiga)
possible_paths = [
    '/content/drive/MyDrive/TP_Final_Vision',
    '/content/drive/My Drive/TP_Final_Vision',
    '/content/drive/Shareddrives/TP_Final_Vision'
]

# detecta la primera que exista
for p in possible_paths:
    if Path(p).exists():
        sys.path.append(p)
        print(f"Módulo cargado desde: {p}")
        break
else:
    raise FileNotFoundError("No se encontró la carpeta TP_Final_Vision en Drive ni Shared Drives.")

from utils_IT import find_root, draw_yolo_boxes

try:
    from google.colab import drive
    drive.mount('/content/drive')
except:
    pass

from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

CLASSES = ['Horse', 'Deer', 'Cow']
ROOTS = [Path('/content/drive/MyDrive'), Path('/content/drive/My Drive'), Path('/content/drive/Shareddrives')]

# Buscar carpeta raíz (usando la función del módulo)
DATASET_ROOT = find_root(ROOTS, CLASSES)
print("DATASET:", DATASET_ROOT)

# Cargar imágenes
IMG_EXTS = {'.jpg','.jpeg','.png','.JPG','.JPEG','.PNG'}
images = {c:[p for p in (DATASET_ROOT/c).rglob('*') if p.suffix in IMG_EXTS] for c in CLASSES}

# Mostrar resumen e imágenes + cajas YOLO
for c, lst in images.items():
    print(f"{c}: {len(lst)} imágenes")
    for p in lst[:2]:
        draw_yolo_boxes(p, CLASSES)
```
