```python
from ultralytics import YOLO

model_rgb = YOLO("yolov8n.pt")

results = model_rgb.train(
    data="data/format_rgb/wildlife_rgb.yaml",
    epochs=20,      
    imgsz=614,        
    batch=6,         
    workers=0,          
    device="cpu",
    name="wildlife_rgb_v8n_cpu"
)
```
