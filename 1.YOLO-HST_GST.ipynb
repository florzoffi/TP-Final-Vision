```python
from src.utils_HST_GST import *
from pathlib import Path
from ultralytics import YOLO
```

```python
ROOT = Path("data")
RGB_ROOT = ROOT / "format_rgb"
T_ROOT   = ROOT / "format_t"

HST_ROOT = ROOT / "format_hst"
GST_ROOT = ROOT / "format_gst"

create_hst_gst_from_roots(
    root_rgb=RGB_ROOT,
    root_t=T_ROOT,
    out_hst_root=HST_ROOT,
    out_gst_root=GST_ROOT,
)
```

```python
# YAML HST
yaml_hst_path = HST_ROOT / "wildlife_hst.yaml"
yaml_hst_content = f"""path: {HST_ROOT}

train: images/train
val: images/val
test: images/test

names:
  0: Cow
  1: Deer
  2: Horse
"""
yaml_hst_path.write_text(yaml_hst_content)
print("YAML HST ->", yaml_hst_path.resolve())

# YAML GST
yaml_gst_path = GST_ROOT / "wildlife_gst.yaml"
yaml_gst_content = f"""path: {GST_ROOT}

train: images/train
val: images/val
test: images/test

names:
  0: Cow
  1: Deer
  2: Horse
"""
yaml_gst_path.write_text(yaml_gst_content)
print("YAML GST ->", yaml_gst_path.resolve())
```

```python
ROOT = Path("data")

YAML_HST = ROOT / "format_hst" / "wildlife_hst.yaml"
YAML_GST = ROOT / "format_gst" / "wildlife_gst.yaml"

# Hiperpar√°metros comunes (ajustables)
COMMON_TRAIN_ARGS = dict(
    imgsz=50,          # m√°s grande por objetos peque√±os
    epochs=5,          # pod√©s subir a 100 si el tiempo da
    batch=2,            # si te da memoria, subilo a 16
    device="cpu",       # est√°s en Mac Intel ‚Üí CPU
    project="runs_wildlife",  # carpeta root para todos los experimentos
    pretrained=True,    # usar pesos pre-entrenados
)
```

```python
# YOLO HST
model_hst = YOLO("yolov8n.pt")
model_hst.train(
    data=str(YAML_HST),
    name="yolo_hst",
    **COMMON_TRAIN_ARGS,
)

# YOLO GST
model_gst = YOLO("yolov8n.pt")
model_gst.train(
    data=str(YAML_GST),
    name="yolo_gst",
    **COMMON_TRAIN_ARGS,
)
```

```python
best_hst = YOLO("runs_wildlife/detect/yolo_hst/weights/best.pt")
best_gst = YOLO("runs_wildlife/detect/yolo_gst/weights/best.pt")

metrics_hst = best_hst.val(data=str(YAML_HST), split="test")
metrics_gst = best_gst.val(data=str(YAML_GST), split="test")

print("HST  - mAP50:",  metrics_hst.box.map50,  "mAP50-95:", metrics_hst.box.map)
print("GST  - mAP50:",  metrics_gst.box.map50,  "mAP50-95:", metrics_gst.box.map)
```
